```python
import numpy as np 
import pandas as pd
# import seaborn as sns
# import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("../datasources/teleco_market_basket.csv")
```

```python
data.head(5)
```

```python
data.shape
```

```python
# keep only rows with items
data = data[data['Item01'].notna()]
data.shape
```

```python
# Convert DF into list of lists
rows = []
for i in range (0, 7501):
    rows.append([str(data.values[i,j]) 
for j in range (0,20)])
```

```python
DE = TransactionEncoder()
array = DE.fit(rows).transform(rows)

transaction_df = pd.DataFrame(array, columns = DE.columns_)
transaction_df
```

```python
cleaned_df = transaction_df.drop(['nan'], axis=1)
cleaned_df.shape
```

```python
# Write normalized data to file
cleaned_df.to_csv("../datasources/D212_3_MBA_df_cleaned.csv", index = False)
```

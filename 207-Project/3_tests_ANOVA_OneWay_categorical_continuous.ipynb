# One-Way ANOVA
## Used to verify how the factor affects the sample mean

H0: A_µ != B_µ != C_µ

H1: NO statistic significant difference between groups

    
    If Statistic >= Critical Value: significant result, reject null hypothesis (H0).
    If Statistic < Critical Value: not significant result, fail to reject null hypothesis (H0).
    
    If p-value <= alpha: significant result, reject null hypothesis (H0).
    If p-value > alpha: not significant result, fail to reject null hypothesis (H0).

```python
categorical_columns = [
                  "Churn",
                  "Generation",
                 "State",
                 "Area",
                 "Marital",
                 "Gender",
                 "Techie",
                 "InternetService",
                 "Multiple",
                 "OnlineBackup",
                 "DeviceProtection",
                 "StreamingTV",
                 "StreamingMovies",
                 "Port_modem",
                 "Tablet",
                 "OnlineSecurity",
                 "TechSupport",
                 "Contract",
                 "PaperlessBilling",
                 "PaymentMethod",
                 "Item1",
                 "Item2",
                 "Item3"
                 ]

continuous_columns = [
                  "Income",
                  "Tenure",
                  "Outage_sec_perweek",
                  "MonthlyCharge",
                  "Bandwidth_GB_Year",
                                  ]
#Filter columns
# categorical_df = df.loc[:, categorical_columns]
# continuous_df = df.loc[:, continuous_columns]
```

# ANOVA sample

```python
# Predictor Factor -> Target
# Generation -> Bandwidth_GB_Year

# Filter data by Factor
boomers_consumption = df[df.Generation == "Boomers"].Bandwidth_GB_Year
gerX_consumption =  df[df.Generation == "GerX"].Bandwidth_GB_Year
gerZ_consumption =  df[df.Generation == "GerZ"].Bandwidth_GB_Year
millennials_consumption =  df[df.Generation == "Millennials"].Bandwidth_GB_Year
postWar_consumption =  df[df.Generation == "Post War"].Bandwidth_GB_Year

# Plot density graph 
import plotnine as p9
print(
    p9.ggplot(df) +
    p9.aes(x='Bandwidth_GB_Year', fill='Generation') +
    p9.geom_density(alpha=0.05)
)

# Run One-Way ANOVA
from scipy import stats
anova = stats.f_oneway(boomers_consumption, gerX_consumption, gerZ_consumption, millennials_consumption, postWar_consumption)
print(anova)
print('Reject H0. Means are the same.')
```

# All variables ANOVA analysis

```python
import numpy as np
unique_values = np.unique(df.Generation)
print(unique_values)
```

```python
from numpy.ma.core import array
from scipy import stats
import numpy as np
import plotnine as p9

myDictionaty = dict()
# Compare every feature with each other
for index, categorical_column_name in enumerate(categorical_columns):
  for index, continuous_column_name in enumerate(continuous_columns):
    features = categorical_column_name + " & " + continuous_column_name
    print("\n", features)

    try:
      categorical_column_values = df.loc[:, categorical_column_name]
      unique_values = np.unique(categorical_column_values)

      array_list = []
      for value in unique_values:
        series_array = df[categorical_column_values == value].loc[:, continuous_column_name]
        # print(array)
        array_list.append(series_array.tolist())
                
      #Perform One-way ANOVA
      # H0: A_µ != B_µ != C_µ
      # H1: NO statistic significant difference between groups
      anova = stats.f_oneway(*array_list)
      statistic, pvalue = anova
      print("ANOVA One-way:", anova)

      #Interpret
      alpha = 0.05
      #P-value is the probability of H0 being true
      if pvalue <= alpha:
        h0 = True
        print('Fail to reject H0. Means are different.')
        key = features
        myDictionaty[key] = anova
      else:
        h0 = False
        print('Reject H0. Means are the same.')

      if(h0):
        # Plot density graph 
        print(
            p9.ggplot(df) +
            p9.aes(x=continuous_column_name, fill=categorical_column_name) +
            p9.geom_density(alpha=0.05)
        )

    except:
      print(features, "FAILLED")

print("\nSUMMARY: \nDifference exists beetween these features")
myDictionaty_sorted = dict(sorted(myDictionaty.items(), key=lambda item: item[1]))
# Print dictionary as json format
import json
print(json.dumps(myDictionaty_sorted, indent=4))
```

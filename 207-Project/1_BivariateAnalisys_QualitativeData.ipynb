# Load dataset

```python
from operator import index
# import into data frame
import pandas as pd
churn = "https://raw.githubusercontent.com/dayanaviana/WGU/main/datasources/churn_clean.csv"
df = pd.read_csv(churn, index_col=0)

df_churn_yes = df[df["Churn"]=="Yes"]
df_churn_no = df[df["Churn"]=="No"]

categorical_columns_array = [
                #  "Churn",
                #  "City",
                #  "State",
                #  "County",
                 "Area",
                 "Marital",
                 "Gender",
                 "Techie",
                 "InternetService",
                 "Multiple",
                 "OnlineBackup",
                 "DeviceProtection",
                 "StreamingTV",
                 "StreamingMovies",
                 "Port_modem",
                 "Tablet",
                 "OnlineSecurity",
                 "TechSupport",
                 "Contract",
                 "PaperlessBilling",
                 "PaymentMethod",
                #  "TimeZone",
                 "Item1",
                 "Item2",
                 "Item3",
                 "Item4",
                 "Item5",
                 "Item6",
                 "Item7",
                 "Item8",
                 "Job",
                 ]
```

# Descriptive Statistics
Describe and summarize data

```python
import matplotlib.pyplot as plt

for column_name in categorical_columns_array:
  print("\n")
  print(column_name)
  counts = df[column_name].value_counts(dropna=False)
  percentages = df[column_name].value_counts(dropna=False, normalize=True)

  counts_yes = df_churn_yes[column_name].value_counts(dropna=False)
  percentages_yes = df_churn_yes[column_name].value_counts(dropna=False, normalize=True)

  counts_no = df_churn_no[column_name].value_counts(dropna=False)
  percentages_no = df_churn_no[column_name].value_counts(dropna=False, normalize=True)
  
  summary = pd.concat([counts, 
                       percentages*100, 
                       counts_yes, 
                       percentages_yes*100, 
                       counts_no, 
                       percentages_no*100
                       ], 
                      axis=1, 
                      keys=['Counts', '%', '| YES:  Counts', '%', '|NO: Counts', '%',]
                      )
  print(summary)
  print("\n")

  # Histogram and show plot 
  # Plot size: 1200px x 400px
  plt.subplots(figsize=(12, 4))
  # configure supplot(nrows,ncols,nsubplots) 
  plt.subplot(1,3,1) 
  plt.hist(df[column_name]) 
  plt.title(column_name)
  plt.subplot(1,3,2) 
  plt.hist(df_churn_yes[column_name]) 
  plt.title("Churn=YES")
  plt.subplot(1,3,3) 
  plt.hist(df_churn_no[column_name]) 
  plt.title("Churn=NO")

  # Bar Graph
  groups = ['Churn', column_name]
  x = groups[1]
  subgroup = groups[0]

  df_pivot = df.groupby(groups).size().reset_index().pivot(columns=subgroup, index=x, values=0)
  print(df_pivot)

  # set width of bar
  barWidth = 0.4

  my_graph = df_pivot.plot.bar( 
                      stacked=True,
                      color=['lightseagreen','tomato'], 
                      figsize=(8,6))

  my_graph.set_title('Quantity of Churn by '+ column_name)
  # my_graph.set_ylim(0,10000)
  my_graph.set_xticklabels(df_pivot.index, rotation=0)


  plt.show() 
```

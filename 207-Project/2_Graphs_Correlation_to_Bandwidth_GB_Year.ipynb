## Correleation (scatterplots)

```python
import matplotlib.pyplot as plt 

columns_array = ["Income", 
                 "Tenure", 
                 "Outage_sec_perweek", 
                 "MonthlyCharge", 
                 "Bandwidth_GB_Year",
                 "Population",
                 "Children",
                 "Age",
                 "Yearly_equip_failure",
                 "Email",
                 "Contacts",
                 "Churn",
                 "City",
                 "State",
                 "County",
                 "Area",
                 "Marital",
                 "Gender",
                 "Techie",
                 "InternetService",
                 "Multiple",
                 "OnlineBackup",
                 "DeviceProtection",
                 "StreamingTV",
                 "StreamingMovies",
                 "Port_modem",
                 "Tablet",
                 "OnlineSecurity",
                 "TechSupport",
                 "Contract",
                 "PaperlessBilling",
                 "PaymentMethod",
                 "TimeZone",
                 "Item1",
                 "Item2",
                 "Item3",
                 "Item4",
                 "Item5",
                 "Item6",
                 "Item7",
                 "Item8",
                 "Job"
                 ]

for column_name in columns_array:
  print("\n")
  
  # Scatter plot
  # Plot size: 1000px x 600px
  plt.subplots(figsize=(10, 6))
  plt.scatter(df[column_name], df["Bandwidth_GB_Year"])
  # Put the x-axis on a logarithmic scale
  # plt.xscale('log')
  #A correlation will become clear when you display the x-axis on a logarithmic scale.
  # plt.xscale('log')
  # Show plot
  plt.title(column_name + ' VS Bandwidth_GB_Year')
  plt.xlabel(column_name)
  plt.ylabel('Bandwidth_GB_Year')
  plt.grid()
  plt.show()
  plt.clf() #clear current figure
```

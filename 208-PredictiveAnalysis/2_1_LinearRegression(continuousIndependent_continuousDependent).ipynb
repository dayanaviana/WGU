#Import Data

```python
import pandas as pd

# import into data frame
churn_data = "https://raw.githubusercontent.com/dayanaviana/WGU/main/datasources/medical_clean.csv"
df = pd.read_csv(churn_data, index_col=0)

continuous_columns = [
                    "Income",
  "VitD_levels",
  "Initial_days",
  # "TotalCharge",
  "Additional_charges",
  "Population",
  "Children",
  "Age",
  "Doc_visits",
  "Full_meals_eaten",
  "vitD_supp",
                                  ]
```

#Linear Regression


```python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols

for column_name2 in continuous_columns:
  column_name1 = "TotalCharge"
  # for column_name2 in continuous_columns:
  print("\n")
  print(column_name1, " & ", column_name2)

  #Correlation - Descriptive statistics
  correlation = df[column_name1].corr(df[column_name2])
  print("Correlation: ", "{:.4f}".format(correlation))

  #Running a linear regression model
  # y = intercept + slope * x
  formula = column_name1 + " ~ " + column_name2
  model = ols(formula, data=df)
  model = model.fit()
  intercept, slope = model.params
  print("Slope:", slope, " Intercept:", intercept)

  # Visualizing pairs of variable with a treand line
  sns.regplot(
      x=column_name2,
      y=column_name1,
      data=df,
      ci=None,
      scatter_kws={"color": "gray"}, 
      line_kws={"color": "black"}
  )
  plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from PyUtils import MyPyUtils as utils

# import into data frame
filePath = "churn_D208-2.csv"
df = pd.read_csv(filePath, index_col=0)
df.columns
```

```python
myList = ['Rural', 'Churn', 'Techie', 'Contract_Monthly', 'Port_modem', 'Tablet',
       'InternetService_Fiber', 'InternetService_DSL', 'Phone', 'Multiple',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'PaperlessBilling','PaymentMethod_Auto',
       ]
for column_name in myList:
  df[column_name] = df[column_name].astype('object')
```

```python
numeric_columns_array = df.select_dtypes(include=['int64','float64','uint8']).columns.to_list()
numeric_columns_array
```

```python
X = df['Churn']
for column_name in numeric_columns_array:
  print("\n")
  print(column_name)
  counts = X.value_counts(dropna=False)
  percentages = X.value_counts(dropna=False, normalize=True)

  sns.boxplot(x = X,
            y = df[column_name])
  plt.show() 
```

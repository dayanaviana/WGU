```python
#imports
import pandas as pd
import matplotlib.pyplot as plt 
from PyRepo import MyPyRepo as repo
df = repo.df
```

```python
df.info()
```

```python
numericc_columns = df.select_dtypes(include=['int64','float64']).columns.to_list()
numericc_columns
```

```python

pd.plotting.scatter_matrix(df, figsize=[20,20], s=10, 
                           marker='o'
                           )
```

```python
for column_x in numericc_columns:
       x = df[column_x]
       y = df["Churn"]
       plt.title('Scatter Plot of Numeric Variables')
       # plt.plot(x, y, ',', color = 'red')
       # plt.xlabel(x.name)
       # plt.ylabel(y.name)
       plt.scatter(y, x, c=x, cmap='rainbow')
       plt.xlabel(y.name)
       plt.ylabel(x.name)
       plt.show()
```
